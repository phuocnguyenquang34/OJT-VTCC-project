```python
import pandas as pd
import CrawlNews
import LDA
import LSTM_model
import Utils
import gensim
```

```python
# Get news
print('Getting news...')
titles, links, contents = CrawlNews.crawl_news()
```

```python
# Preprocess news
print('Preprocessing news...')
contents = Utils.list_preprocessing(contents)
df = pd.DataFrame(list(zip(titles, links, contents)), columns=['Title', 'Link', 'Content'])
df = df.dropna()
```

```python
# Classify news
print('Classifying news...')
# tokenizer = AutoTokenizer.from_pretrained("vinai/phobert-base")
# phoBert = PhoBertModel._load_PhoBert_model()
classes = ['Other',
           'du-lich_diem-den',
           'giai-tri_gioi-sao',
           'giao-duc_tin-tuc',
           'khoa-hoc_tin-tuc',
           'suc-khoe_cac-benh',
           'suc-khoe_tin-tuc',
           'the-thao_bong-da',
           'the-thao_cac-mon-khac',
           'the-thao_world-cup-2022']
lstm = LSTM_model.lstm()
df['Topic'] = df['Content'].apply(lambda x: lstm.predict_text(text=x, classes=classes))
indexOther = df[df['Topic'] == 'Other'].index
df.drop(indexOther , inplace=True)
```

```python
df
```

```python
def topic_modeling(datanews=[], corpus_tfidf=[], dictionary=dict, bow_corpus=[]):
    if len(datanews) < 5: num_topics = 1
    else: num_topics = round(len(datanews) / 5)
    lda_model = gensim.models.LdaMulticore(corpus_tfidf, num_topics=num_topics, id2word=dictionary, passes=10, workers=2)
    topic_dict = dict()
    for idx, topic in lda_model.show_topics(formatted=False, num_words=100):
        words = '|'.join([w[0] for w in topic])
        topic_dict[idx] = words
    topic_modeling, score = [], []
    for idx, news in enumerate(datanews):
        score.append(lda_model[bow_corpus[idx]][0][1])
        topic_modeling.append(topic_dict.get(lda_model[bow_corpus[idx]][0][0]))
    return (topic_modeling, score)
```

```python
#Topic modeling
available_topic = df['Topic'].unique()
count = 1
print('Available topic:')
for topic in available_topic:
    print(str(count) + '. ' + topic)
    count += 1
choice = int(input('Your choice: ')) - 1
```

```python
topic_df = df[df['Topic'] == available_topic[choice]]
datanews = topic_df['Content']
datanews = datanews.apply(lambda x: x.split()).tolist()
dictionary = gensim.corpora.Dictionary(datanews)
bow_corpus = [dictionary.doc2bow(doc) for doc in datanews]
tfidf = gensim.models.TfidfModel(bow_corpus)
corpus_tfidf = tfidf[bow_corpus]
topic_df['Topic modeling'], topic_df['Score'] = topic_modeling(datanews,corpus_tfidf,dictionary,bow_corpus)
```

```python
for topic_model in topic_df['Topic modeling'].unique():
    print(topic_model)
    top4_df = topic_df[topic_df['Topic modeling'] == topic_model].sort_values(by=['Score'], ascending=False)
    for i in range(len(top4_df)):
        print(top4_df['Title'].tolist()[i] + ' ' + top4_df['Link'].tolist()[i])
    print()
```
